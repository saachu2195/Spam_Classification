```python
## Importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Importng the data

spam_data = pd.read_csv("C:/Users/square/Desktop/Machine_Learning_Projects/Spam_Classification/spam.csv", encoding='latin-1')
spam_data.head()
```

```python
## Getting the info of the data

spam_data.info()
```

```python
## Firstly dropping the last 3 variable.

spam_data.drop(columns = ['Unnamed: 2', "Unnamed: 3", "Unnamed: 4"], inplace = True)
spam_data.head()
```

```python
## Checking for any NAN value

spam_data.isnull().sum()
```

Here, Lets assign a value of a ham = 0 which means it is 'Not Spam' & spam = 1 which means it is 'Spam'.
And changing the 'v1' to labels & changing 'v2' to messages.

```python
## Changing the columns name.

spam_data.rename(columns = {'v1': 'labels', 'v2': 'messages'}, inplace = True)
```

```python
## Viewing the data

spam_data.head()
```

```python
## Viewing the types of response.

spam_data['labels'].value_counts()
```

```python
## Getting the percentage of each repsonse.

ham = len(spam_data[spam_data['labels'] == 'ham'])/len(spam_data) * 100
print("% of ham repsonse:", ham)

spam = len(spam_data[spam_data['labels'] == 'spam'])/len(spam_data) * 100
print("% of spam response: ", spam)
```

Since 'ham' response is dominating the reponse 'spam'. May be there can be a chance of Class Imbalance problems.

```python
## Describe

spam_data.describe()
```

```python
## Getting the len of the text of messages variable

spam_data['length'] = spam_data['messages'].apply(len)
spam_data.head()
```

```python
## MApping the value of 'ham' & 'spam' into integer.

spam_data['labels'] = spam_data.labels.map({'ham': 0 , 'spam' : 1})
spam_data.head()
```

```python
## Separating  the variable data into X & Y.

X = spam_data.messages
Y = spam_data.labels

print(X.shape)
print(Y.shape)
```

```python
## Dividing the data into training and testing.

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, random_state = 42)

print("X_train : ", X_train.shape)
print("Y_train : ", Y_train.shape)
print("X_test :  ", X_test.shape)
print("Y_test :  ", Y_test.shape)
```

```python
## CountVectorization

from sklearn.feature_extraction.text import CountVectorizer

## Vectorizing the text.
vect = CountVectorizer(stop_words='english')
vect
```

```python
## Fitting on training data

vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
## Transforming the data.

X_train_transformed = vect.transform(X_train)
X_test_transformed = vect.transform(X_test)
```

```python
## Importing the Naive Bayes library & using MultinomialNB

from sklearn.naive_bayes import MultinomialNB

## Initializing the NB
nb = MultinomialNB()

## Fitting the models
nb.fit(X_train_transformed,Y_train)
```

```python
## Prediction

y_test_predict = nb.predict(X_test_transformed)


## prediction probability
Y_test_proab = nb.predict_proba(X_test_transformed)


## Accuracy
from sklearn import metrics
acc = metrics.accuracy_score(y_test_predict, Y_test)

print("Prediction On Test Data:", acc)
```

```python
## Classification Report

print(metrics.classification_report(y_test_predict, Y_test))
```

```python
## Confusion Matrix

metrics.confusion_matrix(y_test_predict, Y_test)
```

```python
## Persist the model for future use.

from sklearn.externals import joblib
joblib.dump(nb, 'NB_spam_model.pkl')
```
